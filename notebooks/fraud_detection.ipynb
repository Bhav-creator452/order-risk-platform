### Mapping rows of the dataset to the risk engine 

```python
from models import Order , Customer
from scorer import RiskScorer

from rules import (
    HighAmount,
    AccountAge,
    CountryMismatch,
    LargeItemCount
)

rules = [
    HighAmount(threshold=1000),   
    AccountAge(),
    CountryMismatch(),
    LargeItemCount()
]
scorer = RiskScorer(rules)

def row_to_order(row):
        customer = Customer(
            id=row["Customer ID"],
            country="NA",
            account_age_days=row["Account Age Days"],
            #total_past_orders=0,
        )
        return Order(
    id=str(row["Transaction ID"]),
    amount=float(row["Transaction Amount"]),
    currency="USD",          # Placeholder
    customer=customer,
    item_count=int(row["Quantity"]),
    shipping_country="NA"
)
```

### Calculating the number of rows our risk engine fails to map

```python
failed_rows = 0

def predict_order(row):
    global failed_rows

    try:
        order = row_to_order(row)
        score, label = scorer.score(order)
        return label

    except Exception as e:
        failed_rows += 1
        print(f"Row {row.name} failed: {e}")
        return None
```

#### Applying the label mapping to each row 


```python
df["predicted_label"] = df.apply(
    predict_order,
    axis=1
)
```

# Baseline Results

The Week 1 rule-based fraud scorer was evaluated against the fraud dataset.

Evaluation metrics computed:

- Confusion Matrix
- Accuracy
- Precision
- Recall


The rule engine achieved limited recall because several of the original rules could not be fully applied to this dataset. The **NewCustomer** rule was disabled because the dataset does not contain reliable customer purchase history, making it impossible to determine whether a customer is truly new. Additionally, **CountryMismatch** was effectively inactive because the dataset does not provide customer country information, so a placeholder country value was used during the mapping process. As a result, the rule could never detect a country mismatch.

This means that only **three rules** were actively contributing to the final risk score during evaluation. Since the rule engine was operating with fewer signals than originally designed, it identified only a portion of the fraudulent transactions, resulting in a lower recall. This is an expected limitation of the current rule-based baseline rather than an implementation error.

Despite these limitations, the baseline evaluation is valuable because it provides a measurable starting point. Future machine learning models will be trained using all available dataset features and can be directly compared against this rule-based baseline to determine whether they improve fraud detection performance.


## Precision

Precision answers the question:

"Of all the transactions predicted as fraud, how many were actually fraudulent?"

A high precision means the model raises fewer false alarms.

Formula:

Precision = TP / (TP + FP)

---

## Recall

Recall answers the question:

"Of all the actual fraudulent transactions, how many did the model detect?"

A high recall means the model catches most fraud cases.

Formula:

Recall = TP / (TP + FN)

---

## Which metric matters more?

Recall is generally more important for fraud detection because missing fraudulent transactions can lead to financial loss.

Although false positives may inconvenience customers or require manual review, false negatives allow fraudulent transactions to go undetected, making recall the more critical metric.

Why Train/Test Split?

A machine learning model should be evaluated on data it has never seen before.

The dataset is split into:

Training Set (80%): Used to train the model.
Testing Set (20%): Used to evaluate the model.

This provides a realistic estimate of how the model will perform on unseen transactions.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)
```

Why Use Stratification?

Fraud datasets are imbalanced because fraudulent transactions are much rarer than genuine ones.

Using stratify=y ensures that both the training and testing datasets contain approximately the same percentage of fraudulent transactions as the original dataset.

Without stratification, one split might contain too few fraud cases, leading to unreliable evaluation.

## Data Preprocessing

before training a logistic regression model , we need to convert every categorical column into numeric for fitting the data into the model as the model only takes numeric data as an input .

```python
X_train_ml = X_train.copy()
X_test_ml = X_test.copy()
```

## Feature Engineering

It is the process of selecting, transforming, and creating meaningful variables (features) from raw data to help machine learning models learn patterns more effectively. we will transform mostly every column instead of just dropping them 

```python
# converting "transaction date" to datetime format for both training and testing datasets
X_train_ml["Transaction Date"] = pd.to_datetime(X_train_ml["Transaction Date"])
X_test_ml["Transaction Date"] = pd.to_datetime(X_test_ml["Transaction Date"])

# Extracting useful features 
date_features = [
    ("Transaction Month", "month"),
    ("Transaction Day", "day"),
    ("Transaction Weekday", "dayofweek")
]

for new_col, attr in date_features:
    X_train_ml[new_col] = getattr(X_train_ml["Transaction Date"].dt, attr)
    X_test_ml[new_col] = getattr(X_test_ml["Transaction Date"].dt, attr)

X_train_ml["Is Weekend"] = (
    X_train_ml["Transaction Weekday"] >= 5
).astype(int)

X_test_ml["Is Weekend"] = (
    X_test_ml["Transaction Weekday"] >= 5
).astype(int)
```

```python
# Address Match Feature
X_train_ml["Address Match"] = (
    X_train_ml["Shipping Address"]
    ==
    X_train_ml["Billing Address"]
).astype(int)

X_test_ml["Address Match"] = (
    X_test_ml["Shipping Address"]
    ==
    X_test_ml["Billing Address"]
).astype(int)
```

```python
#Removing the original date and address columns as they are no longer needed
columns_to_drop = [
    "Transaction Date",
    "Shipping Address",
    "Billing Address",
    "IP Address" ,      #dropping IP Address as it is not really useful for model 
    "Customer Location"
]

X_train_ml.drop(columns=columns_to_drop, inplace=True, errors="ignore")
X_test_ml.drop(columns=columns_to_drop, inplace=True, errors="ignore")

```

```python

print(X_train_ml.select_dtypes(include="object").columns.tolist())
```

## One hot Encoding categorical features 

```python
categorical_columns = [
    "Payment Method",
    "Product Category",
    "Device Used",
    "Account Age Group"
]

X_train_ml = pd.get_dummies(
    X_train_ml,
    columns=categorical_columns,
    drop_first=True
)

X_test_ml = pd.get_dummies(
    X_test_ml,
    columns=categorical_columns,
    drop_first=True
)
```

```python
#Aligning the columns of training and testing datasets to ensure they have the same featres
X_train_ml, X_test_ml = X_train_ml.align(
    X_test_ml,
    join="left",
    axis=1,
    fill_value=0
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numeric_columns = [
    "Transaction Amount",
    "Customer Age",
    "Quantity",
    "Account Age Days",
    "Transaction Hour",
    "Transaction Month",
    "Transaction Day",
    "Transaction Weekday"
]

X_train_ml[numeric_columns] = scaler.fit_transform(
    X_train_ml[numeric_columns]
)

X_test_ml[numeric_columns] = scaler.transform(
    X_test_ml[numeric_columns]
)
```

## Trade-off Discussion

Fraud detection involves balancing two types of errors:

**False Negative**: A fraudulent transaction is missed. This can result in financial loss, customer dissatisfaction, and reputational damage.

**False Positive**: A legitimate transaction is incorrectly flagged as fraud. This may inconvenience customers by requiring additional verification but usually causes less harm than allowing fraud to go undetected.

Although Logistic Regression identifies more fraudulent transactions, it also produces significantly more false alarms. The Optimized Random Forest provides a better overall trade-off between detecting fraud and minimizing unnecessary alerts, making it the preferred model for this project.
