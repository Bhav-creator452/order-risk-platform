#### Looking for invalid values in dataset

```python
# A Transaction Amount can never be negative
df[df["Transaction Amount"]<0]
```

```python
# A customer age can never be negative
df[df["Customer Age"]<0]
```

##### there is one invalid value in the 'Customer Age' column i.e. -2 and an age can never be negative . Since , it is only one record we will replace it with median age , to preserve the integrity of records and not lose any data point

```python
median_age=df["Customer Age"].median()

# Replacing the invalid value with median age 
df.loc[df["Customer Age"]<0,"Customer Age"]=median_age

df[df["Customer Age"]<0]
```

```python
# Quantity of a product can never be negative
df[df["Quantity"]<0]
```

```python
#Looking for empty or invalid Transaction ID or Customer ID
df[df["Transaction ID"].isna()]

df[df["Customer ID"].isna()]
```

```python
# Label should contain only binary value (0 and 1)

df["Is Fraudulent"].unique()
```

```python
#Account Age can never be negative
df[df["Account Age Days"]<0]
```

```python
# checking for invalid inputs in categorical columns

print(df["Payment Method"].value_counts())

print(df["Device Used"].value_counts())

print(df["Product Category"].value_counts())
```

```python
# Transaction Hour will always be between 0-23
df[(df['Transaction Hour'] < 0) | (df['Transaction Hour'] > 23)]
```

### Exploratory Data Analysis (EDA)

#### first , let's check if any of our data has outliers by plotting a boxplot

```python
#%pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot to check outliers")
plt.show()
```

##### A boxplot is used to identify potential outliers in numerical features. Although Transaction Amount contains several extreme values, these observations will be retained because unusually large transactions indicates fraudulent activity. 

### Getting Insights from charts

```python
avg_amount = df.groupby('Is Fraudulent')['Transaction Amount'].mean()

avg_amount.plot(kind='bar')
plt.title('Average Transaction Amount by Fraud Status')
plt.xlabel('Fraud Status (0 = Not Fraudulent, 1 = Fraudulent)')
plt.ylabel('Average Transaction Amount')
plt.show()
```

##### Fraudulent transactions have an average value of approximately ₹560, compared to about ₹210 for legitimate transactions. This indicates that fraudulent orders tend to involve significantly larger transaction amounts, making transaction amount a potentially important predictor for fraud detection.

```python
hourly_fraud = df.groupby('Transaction Hour')['Is Fraudulent'].mean()

hourly_fraud.plot(kind='line', marker='o')
plt.title('Fraud Rate by Transaction Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Fraud Rate')
plt.show()
```

##### Fraudulent transactions are significantly more common during the early morning hours (12 AM–5 AM). The fraud rate peaks at approximately 12.5% around 5 AM, which is more than four times higher than the average daytime fraud rate of about 3%. This suggests that transaction time is an important indicator of potential fraud.

```python
df['Account Age Group'] = pd.cut(
    df['Account Age Days'],
    bins=[0, 30, 90, 180, 365],
    labels=['0-30', '31-90', '91-180', '181-365']
)

fraud_rate = df.groupby('Account Age Group')['Is Fraudulent'].mean()

fraud_rate.plot(kind='bar')
plt.title('Fraud Rate by Account Age Group')
plt.ylabel('Fraud Rate')
plt.show()
```

##### Newly created accounts (0–30 days old) exhibit a fraud rate of approximately 23%, compared to only about 3% for accounts older than 30 days. This indicates that fraudulent activity is heavily concentrated among new accounts, making account age one of the strongest predictors of fraud in the dataset.

### Getting Insights from Correlation 

```python
sns.heatmap(
    df[['Transaction Amount',
        'Quantity',
        'Customer Age',
        'Account Age Days',
        'Transaction Hour',
        'Is Fraudulent']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```

##### Transaction Amount has the strongest positive relationship with the label i.e. 0.28, risk is higher when transaction amounts are larger too 

##### Account Age days have strong negative correlation of -0.14 with label which means the less the account age is , the more risky the order is .
