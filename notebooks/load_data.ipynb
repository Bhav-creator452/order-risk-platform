#### Looking for invalid values in dataset

```python
# A Transaction Amount can never be negative
df[df["Transaction Amount"]<0]
```

```python
# A customer age can never be negative
df[df["Customer Age"]<0]
```

##### there is one invalid value in the 'Customer Age' column i.e. -2 and an age can never be negative . Since , it is only one record we will replace it with median age , to preserve the integrity of records and not lose any data point

```python
median_age=df["Customer Age"].median()

# Replacing the invalid value with median age 
df.loc[df["Customer Age"]<0,"Customer Age"]=median_age

df[df["Customer Age"]<0]
```

```python
# Quantity of a product can never be negative
df[df["Quantity"]<0]
```

```python
#Looking for empty or invalid Transaction ID or Customer ID
df[df["Transaction ID"].isna()]

df[df["Customer ID"].isna()]
```

```python
# Label should contain only binary value (0 and 1)

df["Is Fraudulent"].unique()
```

```python
#Account Age can never be negative
df[df["Account Age Days"]<0]
```

```python
# checking for invalid inputs in categorical columns

print(df["Payment Method"].value_counts())

print(df["Device Used"].value_counts())

print(df["Product Category"].value_counts())
```

```python
# Transaction Hour will always be between 0-23
df[(df['Transaction Hour'] < 0) | (df['Transaction Hour'] > 23)]
```
